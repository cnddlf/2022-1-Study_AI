#4. 하이퍼 파라미터

```python
from sklearn.model_selection import GridSearchCV

def get_best_hyper_parameter(clf, cost_list, data, label):
  params = {
      'C' : cost_list
  }

  grid_cv = GridSearchCV(clf, param_grid=params, scoring = 'accuracy', cv = 5)
  grid_cv.fit(data, label)
  print('Best Score: ', grid_cv.best_score_)
  print('Best Hyper Parameter: ', grid_cv.best_params_)
```

##4.1 선형 모델 분석

```python
import sklearn.svm as svm
svm_clf = svm.SVC(kernel = 'linear')
cost_list = [0.001, 0.1, 0.1, 1, 10, 25, 50, 100]
```

```python
get_best_hyper_parameter(svm_clf, cost_list, df_data, df_labels)
```

##4.2 비선형 모델 분석

```python
import sklearn.svm as svm
svm_clf = svm.SVC(kernel = 'rbf')
cost_list = [0.001, 0.1, 0.1, 1, 10, 25, 50, 100]
```

```python
get_best_hyper_parameter(svm_clf, cost_list, df_data, df_labels)
```
