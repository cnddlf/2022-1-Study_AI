#1. 데이터 읽기

```python
# load_boston을 이용하여 데이터를 읽어들이고, data와 label을 데이터 프레임으로 만드시오.
from sklearn.datasets import load_boston

# boston 데이터셋 로드
boston = load_boston()
```

```python
import pandas as pd

# boston dataset을 DataFrame으로 전환
df_data = pd.DataFrame(boston.data, columns = boston.feature_names)
# boston dataset의 target array는 주택 가격임. 이를 PRICE 컬럼으로 DataFrame에 추가함.
df_labels = boston.target
```

#2. 모델 만들기

##2.1 훈련 / 검증 데이터 추출

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_data, df_labels, test_size = 0.3, random_state = 62)
```

##2.2 모델 학습 / 정확도

```python
from sklearn.linear_model import LinearRegression

# 모델 생성
lr = LinearRegression()
# 모델 학습
lr.fit(X_train, y_train)
# 정확도 계산
lr.score(X_test, y_test)
```

##2.3 성능 평가

```python
pred = lr.predict(X_test)
pred
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

# 실 데이터와 예측 데이터간의 거리로 수치가 작을 수록 실 데이터와 비슷한 결과를 얻는다는 검증을 하는 것이다.
mse = mean_squared_error(y_test, pred)
print('mse : ', mse)
print('rmse : ', np.sqrt(mse))
```

#3. 교차검증

```python
lr = LinearRegression()
```

```python
from sklearn.model_selection import cross_val_score

neg_mean_squared = cross_val_score(lr, df_data, df_labels, scoring = 'neg_mean_squared_error', cv=5)
neg_mean_squared
```

```python
rmse = np.sqrt(-1*neg_mean_squared)
rmse*-1
```
